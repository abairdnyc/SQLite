```python
import sqlite3
import pandas as pd
con = sqlite3.connect("sakila.db")

#Set function as our sql_to_pandas

def sql_to_df(sql_query):
    #Use pandas to pass sql query using connection from SQLite3
    df = pd.read_sql(sql_query, con)
    
    #Show the resulting DataFrame
    return df
```

```python
#Count the number of customers
query = ''' SELECT COUNT(customer_id) FROM customer; '''

#Grab
sql_to_df(query).head()
```

```python
#First the % wildcard

#Select any customers whos name star with an M
query = '''SELECT * FROM customer WHERE first_name LIKE 'M%' ; '''

#Grab
sql_to_df(query).head()
```

```python
#Next the _wildcrd

#Select any customers whos last name ends with ing
query = '''SELECT * FROM customer WHERE last_name LIKE '_ING' ; '''

#Grab
sql_to_df(query).head()
```

```python
#Finally the [character_list] wildcard

#Select any customers whose first name begins with an A or a B
query = ''' SELECT * FROM customer WHERE first_name GLOB '[AB]*' ; '''

#Grab
sql_to_df(query).head()
```

```python
#Select all customer and order results by last name
query = '''SELECT * FROM customer ORDER BY last_name ; '''

#Grab
sql_to_df(query).head()
```

```python
#Select all customers and order results by last name, DESCENDING
query = ''' SELECT * FROM customer ORDER BY last_name DESC; '''

#Grab
sql_to_df(query).head()
```

```python
#Count the number of customer per store

query = ''' SELECT store_id , COUNT(customer_id) FROM customer GROUP BY store_id; '''

#Grab
sql_to_df(query).head()
```
